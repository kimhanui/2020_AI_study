## 데이터 로더

```
# [1] 데이터 로더
! curl 'https://raw.githubusercontent.com/sejongresearch/2020.Spring.AI/master/HW/cabbage_price.csv' -o 'cabbage_price.csv'
print('Done..\n') 

import torch
import torch.optim as optim
import numpy as np

xy = np.loadtxt('cabbage_price.csv', delimiter=',', dtype=np.float32, skiprows=1, usecols=range(1,6))

x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]  #마지막 열만

x_data = torch.FloatTensor(x_data)
y_data = torch.FloatTensor(y_data)

print(x_data[:5]) #5개만 테스트로 보여줌
print(y_data[:5])
print(x_data.shape, y_data.shape)
```

## 모델 학습

```
# [2]모델 학습
W = torch.zeros((4, 1), requires_grad=True)       # ------- W 열크기 왜 1인가?
b = torch.zeros(1, requires_grad=True)
#optimizer 설정
optimizer = optim.SGD([W,b], lr=0.00003)

nb_epochs =1700000
for epoch in range(nb_epochs+1):

    #hypothesis
    hypothesis = x_data.matmul(W) + b

    #cost계산(수렴하면 됨. 꼭 0인건 아님)
    cost = torch.mean((hypothesis - y_data) ** 2)
    #cost->hypothesis 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    #10000번째마다 출력
    if epoch % 10000 == 1:
        print('Epoch: {:4d}/{} cost:{:.6f} hypothesis:{}'.format(epoch, nb_epochs, cost.item(),
                                                                   hypothesis.detach()))
print('Epoch: {:4d}/{} cost:{:.6f} hypothesis:{}'.format(epoch, nb_epochs, cost.item(),
                                                                   hypothesis.detach()))
```

## 정답 예측

```
# [3] 정답 예측
x_test = torch.FloatTensor([-2.7, -6.6, 2.0, 0.1])
predict = x_test.matmul(W) + b
print('predict: {:.4f}'.format(predict.item()))
```
